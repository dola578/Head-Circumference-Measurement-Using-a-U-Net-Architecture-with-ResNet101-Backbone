```python
import os
import pywt
import numpy as np
import cv2

def improved_wavelet_denoise(image, wavelet='coif4', level=2):
    """
    Enhanced wavelet denoising with proper coefficient handling
    """
    if image is None:
        raise ValueError("Image not loaded")
    
    # Convert to grayscale if needed
    if len(image.shape) == 3:
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Pad image to reduce edge artifacts
    pad_size = 2**level
    image = cv2.copyMakeBorder(image, pad_size, pad_size, pad_size, pad_size, 
                             cv2.BORDER_REFLECT)
    
    # Wavelet decomposition
    coeffs = pywt.wavedec2(image, wavelet=wavelet, level=level)
    
    # Adaptive thresholding - PROPERLY HANDLE COEFFICIENTS
    threshold = 0.5 * np.median(np.abs(coeffs[-1]))
    
    # Process coefficients while maintaining structure
    new_coeffs = [coeffs[0]]  # Keep approximation coefficients
    for i in range(1, len(coeffs)):
        # Process each detail coefficient tuple (horizontal, vertical, diagonal)
        processed = []
        for c in coeffs[i]:
            if c is not None:
                processed.append(pywt.threshold(c, value=threshold, mode='soft'))
            else:
                processed.append(None)
        new_coeffs.append(tuple(processed))
    
    # Reconstruction
    reconstructed = pywt.waverec2(new_coeffs, wavelet=wavelet)
    
    # Remove padding and clip values
    reconstructed = reconstructed[pad_size:-pad_size, pad_size:-pad_size]
    return np.clip(reconstructed, 0, 255).astype(np.uint8)

def edge_preserving_smoothing(image):
    """
    Replace Gaussian Blur with better alternatives:
    - Bilateral filter (best for edge preservation)
    - Non-local Means (best for speckle)
    """
    # Option 1: Bilateral Filter (balanced)
    smoothed = cv2.bilateralFilter(image, d=9, sigmaColor=75, sigmaSpace=75)
    return smoothed

def process_images(input_dir, output_dir):
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    for filename in os.listdir(input_dir):
        if (filename.lower().endswith(('.png', '.jpg', '.jpeg')) and 
            'annotation' not in filename.lower()):
            
            image_path = os.path.join(input_dir, filename)
            image = cv2.imread(image_path, cv2.IMREAD_UNCHANGED)

            if image is not None:
                # Step 1: Improved wavelet denoising
                denoised = improved_wavelet_denoise(image)
                
                # Step 2: Edge-preserving smoothing
                smoothed = edge_preserving_smoothing(denoised)
                
                output_path = os.path.join(output_dir, filename)
                cv2.imwrite(output_path, smoothed)
                print(f"Processed {filename}")
            else:
                print(f"Failed to load {filename}")

# Usage
process_images('training_set/training_set', 'training_set/denoised_training_set')
```
