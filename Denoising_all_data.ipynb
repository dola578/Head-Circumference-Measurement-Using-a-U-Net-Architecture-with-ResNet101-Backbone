```python
import os
import pywt
import numpy as np
import cv2

def improved_wavelet_denoise(image, wavelet='coif4', level=2):
    """
    Enhanced wavelet denoising with proper coefficient handling
    """
    if image is None:
        raise ValueError("Image not loaded")

    # Convert to grayscale if needed
    if len(image.shape) == 3:
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Pad image to reduce edge artifacts
    pad_size = 2**level
    image = cv2.copyMakeBorder(image, pad_size, pad_size, pad_size, pad_size,
                             cv2.BORDER_REFLECT)

    # Wavelet decomposition
    coeffs = pywt.wavedec2(image, wavelet=wavelet, level=level)

    # Adaptive thresholding - PROPERLY HANDLE COEFFICIENTS
    threshold = 0.5 * np.median(np.abs(coeffs[-1]))

    # Process coefficients while maintaining structure
    new_coeffs = [coeffs[0]]  # Keep approximation coefficients
    for i in range(1, len(coeffs)):
        # Process each detail coefficient tuple (horizontal, vertical, diagonal)
        processed = []
        for c in coeffs[i]:
            if c is not None:
                processed.append(pywt.threshold(c, value=threshold, mode='soft'))
            else:
                processed.append(None)
        new_coeffs.append(tuple(processed))

    # Reconstruction
    reconstructed = pywt.waverec2(new_coeffs, wavelet=wavelet)

    # Remove padding and clip values
    reconstructed = reconstructed[pad_size:-pad_size, pad_size:-pad_size]
    return np.clip(reconstructed, 0, 255).astype(np.uint8)

def edge_preserving_smoothing(image):
    """
    Replace Gaussian Blur with better alternatives:
    - Bilateral filter (best for edge preservation)
    - Non-local Means (best for speckle)
    """
    # Option 1: Bilateral Filter (balanced)
    smoothed = cv2.bilateralFilter(image, d=9, sigmaColor=75, sigmaSpace=75)
    return smoothed

def process_images(input_dir, output_dir):
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    for filename in os.listdir(input_dir):
        if (filename.lower().endswith(('.png', '.jpg', '.jpeg')) and
            'annotation' not in filename.lower()):

            image_path = os.path.join(input_dir, filename)
            image = cv2.imread(image_path, cv2.IMREAD_UNCHANGED)

            if image is not None:
                # Step 1: Improved wavelet denoising
                denoised = improved_wavelet_denoise(image)

                # Step 2: Edge-preserving smoothing
                smoothed = edge_preserving_smoothing(denoised)

                output_path = os.path.join(output_dir, filename)
                cv2.imwrite(output_path, smoothed)
                print(f"Processed {filename}")
            else:
                print(f"Failed to load {filename}")

# Usage
process_images('training_set/training_set', 'training_set/denoised_training_set')
```

```python
import matplotlib.pyplot as plt
import random
import cv2, os

def visualize_ultrasound_denoising(input_dir, output_dir):
    files = [f for f in os.listdir(input_dir) if f.lower().endswith('.png')]
    f = random.choice(files)
    before = cv2.imread(os.path.join(input_dir, f), cv2.IMREAD_GRAYSCALE)
    after  = cv2.imread(os.path.join(output_dir, f), cv2.IMREAD_GRAYSCALE)

    plt.figure(figsize=(10,5))
    plt.subplot(1,2,1)
    plt.imshow(before, cmap='gray'); plt.title('Noisy');  plt.axis('off')
    plt.subplot(1,2,2)
    plt.imshow(after,  cmap='gray'); plt.title('Denoised'); plt.axis('off')
    plt.suptitle(f'Ultrasound Sample: {f}')
    plt.tight_layout()
    plt.show()

visualize_ultrasound_denoising('training_set/training_set', 'training_set/denoised_training_set')
```

```python
import os
import cv2
import numpy as np
from skimage.filters import sobel
import pandas as pd

def compute_enl(roi):
    """
    Equivalent Number of Looks (ENL) for a homogeneous ROI.
    ENL =  / 
    """
    mu = roi.mean()
    sigma = roi.std(ddof=1)
    return mu**2 / (sigma**2 + 1e-8)

def compute_cnr(roi_obj, roi_bg):
    """
    Contrast-to-Noise Ratio (CNR) between an object ROI and a background ROI.
    CNR = |_obj - _bg| / sqrt(_obj + _bg)
    """
    mu_obj, sigma_obj = roi_obj.mean(), roi_obj.std(ddof=1)
    mu_bg, sigma_bg = roi_bg.mean(), roi_bg.std(ddof=1)
    return abs(mu_obj - mu_bg) / np.sqrt(sigma_obj**2 + sigma_bg**2 + 1e-8)

def compute_epi(img_noisy, img_denoised, edge_thresh=0.1):
    """
    Edge Preservation Index (EPI):
    ratio of gradient magnitudes on edge pixels (Sobel) after vs. before denoising.
    """
    grad_noisy = sobel(img_noisy.astype(float))
    edges = grad_noisy > edge_thresh
    grad_den = sobel(img_denoised.astype(float))
    return grad_den[edges].sum() / (grad_noisy[edges].sum() + 1e-8)

# --- directories ---
input_dir    = 'training_set/training_set'
denoised_dir = 'training_set/denoised_training_set'

results = []

for fname in os.listdir(input_dir):
    if not fname.lower().endswith(('.png', '.jpg', '.jpeg')):
        continue

    # Load as grayscale
    noisy_path = os.path.join(input_dir, fname)
    den_path   = os.path.join(denoised_dir, fname)
    img_noisy  = cv2.imread(noisy_path, cv2.IMREAD_GRAYSCALE)
    img_den    = cv2.imread(den_path,   cv2.IMREAD_GRAYSCALE)

    if img_noisy is None or img_den is None:
        continue

    h, w = img_noisy.shape
    cy, cx = h // 2, w // 2

    # ROI for ENL: central 5050 patch
    roi_enl_noisy = img_noisy[cy-25:cy+25, cx-25:cx+25]
    roi_enl_den   = img_den[cy-25:cy+25,   cx-25:cx+25]

    # ROIs for CNR: one background, one object (reuse ENL patch as object)
    roi_bg_noisy = img_noisy[10:60, 10:60]
    roi_bg_den   = img_den[10:60, 10:60]
    roi_obj_noisy = roi_enl_noisy
    roi_obj_den   = roi_enl_den

    # Compute metrics
    enl_noisy = compute_enl(roi_enl_noisy)
    enl_den   = compute_enl(roi_enl_den)
    cnr_noisy = compute_cnr(roi_obj_noisy, roi_bg_noisy)
    cnr_den   = compute_cnr(roi_obj_den,   roi_bg_den)
    epi       = compute_epi(img_noisy, img_den)

    results.append({
        'filename': fname,
        'ENL_noisy': enl_noisy,
        'ENL_denoised': enl_den,
        'CNR_noisy': cnr_noisy,
        'CNR_denoised': cnr_den,
        'EPI': epi
    })

# Build DataFrame and compute improvements
df = pd.DataFrame(results)
df['ENL'] = df['ENL_denoised'] - df['ENL_noisy']
df['CNR'] = df['CNR_denoised'] - df['CNR_noisy']

# Display results
print(df)
print("\nAverage improvements:")
print(df[['ENL','CNR','EPI']].mean())
```
